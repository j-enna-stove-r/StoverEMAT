# data description

- The dataset contains information about my Spotify stats including my top artist for the month and my top track of the month.
- There will be three rows and up to twelve columns of data for each stat because it will show the month, artist/ track, and the amount of plays for the artist or track.
- The month will be labeled 1-11 regardless of the artist stat or track stat. The artist will go numerical from top to bottom along with the track so it corresponds with the month correctly. The plays will correlate with the track/ artist and how many times I played the artist/ track in that month.

# data with stats

```python
import os
import json
import pandas as pd

folder = "/Users/jenna/Desktop/spotify_data"

records = []

for file in os.listdir(folder):
    if file.endswith(".json"):
        path = os.path.join(folder, file)

        try:
            data = json.load(open(path))
        except:
            continue

        if isinstance(data, dict):
            for key in ["plays", "streams", "items"]:
                if key in data:
                    data = data[key]

        if not isinstance(data, list):
            continue

        for entry in data:
            if not isinstance(entry, dict):
                continue

            ts = entry.get("ts") or entry.get("endTime")
            artist = entry.get("artistName")
            track = entry.get("trackName")

            if ts:
                records.append({
                    "timestamp": ts,
                    "artist": artist,
                    "track": track
                })

df = pd.DataFrame(records)
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
df = df.dropna(subset=["timestamp"])

# Filter 2025
df_2025 = df[df["timestamp"].dt.year == 2025].copy()
df_2025["month"] = df_2025["timestamp"].dt.to_period("M")

print("Loaded rows:", len(df_2025))
```

```python
top_artist_per_month = (
    df_2025.groupby(["month", "artist"])
    .size()
    .reset_index(name="plays")
    .sort_values(["month", "plays"], ascending=[True, False])
    .groupby("month")
    .first()
)

top_artist_per_month
```

```python
top_song_per_month = (
    df_2025.groupby(["month", "track"])
    .size()
    .reset_index(name="plays")
    .sort_values(["month", "plays"], ascending=[True, False])
    .groupby("month")
    .first()
)

top_song_per_month
```
